```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt 
from statsmodels.tsa.stattools import adfuller
```

```python
df=pd.read_csv('C:\\Users\\Admin\\OneDrive\\Desktop\\DESKTOP_FILES\\Internship\\data.csv')
print(df.head())
```

```python
df.columns=['Month','Milk Production']
print(df.head())
print(df.tail())     #we can see here that 168 entry is a NAN value so will drop it
df.drop(168,axis=0,inplace=True)
print(df.tail())
```

```python
df['Month']=pd.to_datetime(df['Month'])
print(df.head())
print(df.tail())
df.set_index(['Month'],inplace=True)  #we need data based on Month so we are making it as an index for easy use 
print(df.head())

df.plot()
```

```python
ts= df['Milk Production']
ts.rolling(12).mean().plot()
```

```python
ts.rolling(12).std().plot()
```

```python
pvalue=adfuller(ts)
print(pvalue)
# # shifting coz pvalue is greater than 0.02
df['Milk Production']=df['Milk Production'].rolling(12).mean().shift(4)
print(df['Milk Production'])
```

```python
df.dropna(inplace=True)                          # since Nan values are added due to shifting
```

```python
df['Milk Production'].head()
```

```python
pvalue=adfuller(df['Milk Production'])
print(pvalue)
```

```python
df['Milk Production']=df['Milk Production'].rolling(12).std().shift(12)
```

```python
df.head()
```

```python
df.dropna(inplace=True)
```

```python
df.head()
```

```python
pvalue=adfuller(df['Milk Production'])
print(pvalue)
```

```python
df.plot()
```

```python
ts2=df['Milk Production']
```

```python
ts2.describe()
```

```python
df['First Diff']=ts2-ts2.shift(1)
```

```python
r=adfuller(df['First Diff'].dropna())
```

```python
r
```

```python
df['Seasonal Diff']=ts2-ts2.shift(12)
```

```python
r=adfuller(df['Seasonal Diff'].dropna())
```

```python
r
```

```python
df['Seasonal first Diff']=df['First Diff']-df['First Diff'].shift(12)
```

```python
r=adfuller(df['Seasonal first Diff'].dropna())
```

```python
r
```

```python
import statsmodels.api as sm 
```

```python
model=sm.tsa.statespace.SARIMAX(df['Milk Production'],order=(0,1,0),seasonal_order=(1,1,1,12))
```

```python
res=model.fit()
```

```python
predict=res.predict(start=150,end=180)
```

```python
predict.plot()
```
